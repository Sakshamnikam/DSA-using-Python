Linear Search

```python
a=[12,34,12,3,1,3,4]

def search(arr,x):

    for i in range(0,len(a)):
        if(arr[i]==x):
            return i
            
    return -1    
    
result=search(a,3)
if(result == -1):
    print("Element is not present in array")
else:
    print("Element is present at index", result)
```

Binary Search

```python
ls=[1,2,3,4,5]

def binary_search(ls,x):
    low=0
    high=len(ls)-1

    while low<=high:
        mid=low+(high-low)//2

        if (ls[mid]==x):
            return mid
        elif(ls[mid]>x):
            low = mid + 1
        else:
            high = mid - 1
    return -1        
    
res=binary_search(ls,5)
if(res<0):
    print('Not found!')
else:
    print(res)
```
